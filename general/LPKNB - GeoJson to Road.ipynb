```python
import os
import json
import re
```

```python
company=Company.objects.get(name='Modera - Consórcio ME2')
```

```python
def convert_road_kartado(a):
    padrao = re.compile(r'\d', re.IGNORECASE)
    resultado = padrao.search(a)
    primeira_ocorrencia = resultado.start()
    primeira_string = a[:primeira_ocorrencia]
    restante = a[primeira_ocorrencia:]
    return primeira_string+'-'+restante
```

```python
new_forms = []

for index, file in enumerate(os.listdir(".")):
    if not '.geojson' in file:
        continue
    file_name = './{}'.format(file)
    print('------')
    print(file_name)
with open(file_name, 'r+',encoding="utf8") as fo:
        form = json.load(fo)
        print(form['name'])
        new_forms.append(form)
```

```python
len(new_forms),new_forms[0]
```

```python
roads={}
for a in new_forms:
    for b in a['features']:
        road_name = b['properties']['Rodovia']
        km_inicial = b['properties']['KmInicial']
        coo_long_inicial = b['geometry']['coordinates'][0][0][0]
        coo_lat_inicial = b['geometry']['coordinates'][0][0][1]
        km_final = b['properties']['KmFinal']
        coo_long_final = b['geometry']['coordinates'][0][-1][0]
        coo_lat_final = b['geometry']['coordinates'][0][-1][1]
        if not roads.get(road_name):
            roads[road_name]={}
        roads[road_name][km_inicial]=[coo_long_inicial,coo_lat_inicial]
        roads[road_name][km_final]=[coo_long_final,coo_lat_final]
        # print(road_name,km_inicial,km_final)
```

```python
roads
```

```python
roads_final=[]
for a in roads:
    chaves_ordenadas = sorted(roads[a].keys())
    roads_ordenada={}
#     print(chaves_ordenadas)
    for chave in chaves_ordenadas:
        if not roads_ordenada.get(a):
            roads_ordenada[a]={}
        roads_ordenada[a][chave] = roads[a][chave]
    roads_final.append(roads_ordenada)
```
