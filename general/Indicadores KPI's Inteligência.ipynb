# Atividades com expectativa de retorno %

```python
months_name=['Dezembro','Janeiro', 'Fevereiro', 'Março' , 'Abril','Maio','Junho','Julho', 'Agosto','Setembro','Outubro','Novembro']
```

# ATIVIDADES REALIZADAS

```python
# months_name=['Maio', 'Junho', 'Julho', 'Agosto','Setembro', 'Outubro', 'Novembro', 'Dezembro','Janeiro', 'Fevereiro', 'Março' , 'Abril']
months=[39,37,51,89,63,72,89,93,107,90,93,144]
```

```python
# fig = plt.figure(figsize=(20,10))
plt.figure(figsize=(20,10))


plt.plot(months_name, months, '-k', marker='o', ms=20, mfc='r', mec='k')
plt.xlabel('Meses', fontsize=20, labelpad=15)
plt.ylabel('Valores', fontsize=20)
plt.title('Atividades realizadas', fontsize=40, pad=40)

axes = plt.gca()
axes.set_ylim([0, 160])

for x, y in zip(months_name, months):
    
    label=y
    plt.annotate(label, (x, y), textcoords="offset points", xytext=(0,-25), ha='center')

# plt.show()
# plt.savefig('realizada_setembro_2022.jpeg', dpi=300)
```

# Tipos de atividades

```python
data_name=['Edição de dados/Datafix', 'Análise/Extração de dados', 'Dúvidas', 'Outros']
data=[68, 17, 2,6]
```

```python
plt.figure(figsize=(30,15))


plt.barh(data_name,data)
plt.xlabel('Quantidade', fontsize=30, labelpad=30)
plt.title('Tipos de atividades', fontsize=40, pad=40)

axes = plt.gca()
# axes.set_ylabel('', fontsize=60)
axes.invert_yaxis()
axes.axes.set_xlim([0, 80])
rects = axes.patches
axes.tick_params(axis='y', which='major', labelsize=20)

for rect, label in zip(rects, data):
    x = rect.get_width()
    y = rect.get_y() + rect.get_height() / 2
    plt.annotate(label, (x, y), textcoords="offset points",  xytext=(20,0), ha='left', va='center', fontsize=15)

# plt.show()
# plt.savefig('contagem_setembro_2022.jpeg', dpi=300)
```

```python
# fig = plt.figure(figsize=(20,10))
plt.figure(figsize=(20,10))


plt.plot(months_name, months, '-k', marker='o', ms=20, mfc='r', mec='k')
plt.xlabel('Meses', fontsize=20, labelpad=15)
plt.ylabel('Valores', fontsize=20)
plt.title('Pontuação perdida', fontsize=40, pad=40)

axes = plt.gca()
axes.set_ylim([-5, 50])

for x, y in zip(months_name, months):
    
    label=y
    plt.annotate(label, (x, y), textcoords="offset points", xytext=(0,-30), ha='center')



# plt.show()
plt.savefig('perdida_setembro_2022.jpeg', dpi=300)
```
