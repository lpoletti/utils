### Remover informação de 'un' dos campos com autofill

```python
from django.db.models import Q
from tqdm.notebook import tqdm
```

```python
company = Company.objects.get(name__icontains="Arteris Litoral Sul - Conserva")
company
```

```python
lista = [74,75,76,77,78,79,80,81,82]
```

```python
occ = OccurrenceType.objects.filter(company=company,occurrence_kind__in=lista)
occ.count()
```

```python
repor = Reporting.objects.filter(occurrence_type__in=occ, company=company)
repor.count()
```

```python
repor.delete()
```

```python
occ.delete()
```

```python
r=Reporting.objects.filter(company=company).exclude(Q(form_data__has_key="origin") & ~Q(form_data__origin=[])).exclude(occurrence_type__occurrence_kind='18')
r.count()
```

```python
user = User.objects.get(username='rlcs')
```

```python
for a in tqdm(r):
    a.form_data['origin'] = ['2']
    a._history_user = user
    a.save()
```
