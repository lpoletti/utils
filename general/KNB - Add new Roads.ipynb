# Criação dos Marks e nome da rodovia para copiar e colar no django

```python
from openpyxl import load_workbook
import json

wb = load_workbook(filename='Kartado - Marco Quilometrico.xlsx')
sheet = wb.active
data = []

headers = [cell.value for cell in sheet[1]]

for row in sheet.iter_rows(min_row=2):
    row_dict = {}
    for key, cell in zip(headers, row):
        row_dict[key] = cell.value

    data.append(row_dict)
```

```python
data[17]
```

```python
values=[]
obj={}
for i,a in enumerate(data[17:23]):
    nome = a['NOME'].split(' - ')[0]
    if nome not in obj:
        obj[nome]={}
    if a['KM'] is not None:
        obj[nome][str(i)]={}
        road_km = str(a['KM']).strip().replace('+','.') if isinstance(a['KM'],str) else 0
        latitude = a['LATITUDE']
        longitude = a['LONGITUDE']
#         lon, lat = transform(inProj, outProj, latitude, longitude)
#         print(float(latitude)*-1,float(longitude)*-1)
        latitude =float(latitude)
        longitude = float(longitude)
        obj[nome][str(i)]['km'] = float(road_km)
        obj[nome][str(i)]['key'] = i
#         obj[str(i)]['mapbox_call'] = True
        obj[nome][str(i)]['point'] ={}
        obj[nome][str(i)]['point']['type'] = "Point"
        obj[nome][str(i)]['point']['coordinates'] = [longitude,latitude]
values.append(obj)
```
