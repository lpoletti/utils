```python
from tqdm.notebook import tqdm
import json
```

```python
company = Company.objects.get(name__icontain="Cart")
company
```

```python
occ = OccurrenceType.objects.filter(company=company)
occ[0].__dict__
```

```python
occ[0].form_fields['fields'][0]['displayName']
```

```python
for a in tqdm(occ):
    a.form_fields = json.loads(json.dumps(a.form_fields, indent=4, ensure_ascii=False).replace('Recebido em:','Recebido em').replace('Solução adotada:','Solução adotada'))
    a.save()    
```
