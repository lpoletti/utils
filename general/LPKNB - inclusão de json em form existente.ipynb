```python
from copy import deepcopy
from tqdm.notebook import tqdm
```

```python
company = Company.objects.get(name="Arteris Via Paulista - Faixa de Domínio")
company
```

```python
occurrence = OccurrenceType.objects.filter(company=company)
occurrence.count()
```

```python
for ac in tqdm(occurrence):
    size = len(ac.form_fields['fields'])
    
    ac.form_fields['fields'].append({
                                        "id": size+1,
                                        "apiName": "startHour",
                                        "dataType": "timestamp",
                                        "displayIf": {
                                            "===": [{
                                                "var": "data.formData.stopAndGo"
                                            }, True]}
                                        })
    ac.form_fields['fields'].append({
                                        "id": size+2,
                                        "apiName": "endHour",
                                        "dataType": "timestamp",
                                        "displayIf": {
                                            "===": [{
                                                "var": "data.formData.stopAndGo"
                                            }, True]}
                                        })
    ac.form_fields['groups'][0]['members'].append(size+1)
    ac.form_fields['groups'][0]['members'].append(size+2)
    ac.save()
```
