```python
def formatar_valores(valores):
    grupos = [valores[i:i+3] for i in range(0, len(valores), 3)]
    grupos_formatados = [f"{grupo[0]} {grupo[1]} {grupo[2]}" for grupo in grupos]
    return ", ".join(grupos_formatados)
```

```python
def road_print(roads):
    key='AIzaSyB0HWaR96SOjW2ujDgd5HbBHTr94m7gvNA'
    
    gmaps=[]
    for i,a in enumerate(roads):
        first_lat = a.path[0][1]
        last_lat = a.path[-1][1]
        first_long = a.path[0][0]
        last_long = a.path[0][0]
        
        if roads[0].path and len(roads[0].path) > 1:
            media_lat = (a.path[0][1] + a.path[-1][1])/2
            media_long = (a.path[0][0] + a.path[-1][0])/2
        elif roads[0].path:
            media_lat = a.path[0][1]
            media_long = a.path[0][0]

        gmap = gmplot.GoogleMapPlotter(media_lat, media_long, 13, apikey=key)

        # Adiciona um marcador no mapa
        gmap.directions(
            (first_lat, first_long),
            (last_lat, last_long),
            waypoints=[
                (p[0],p[1]) for p in a.path
            ]
        )
        gmaps.append(('LINESTRING Z ',[r for p in a.path for r in p]))
        gmap.draw('mapa_'+str(i)+'.html')
    return gmaps
    
```
