```python
from openpyxl import load_workbook

# Carrega a planilha do Excel
wb = load_workbook(filename='Terraplenos_estruturas_contencao.xlsx')
ws = wb.active

# Cria um dicionário vazio para armazenar as informações lidas
data = {}

# Obtém o número de linhas e colunas da planilha
num_rows = ws.max_row
num_cols = ws.max_column

# Itera pelas colunas da planilha
for col in range(1, num_cols + 1):
    # Obtém o valor da célula do cabeçalho da coluna (primeira linha)
    header = ws.cell(row=1, column=col).value

    # Cria uma lista vazia para armazenar os valores da coluna
    column_data = []

    # Itera pelas células da coluna (exceto a primeira linha, que é o cabeçalho)
    for row in range(2, num_rows + 1):
        # Obtém o valor da célula
        cell_value = ws.cell(row=row, column=col).value

        # Adiciona o valor da célula à lista de valores da coluna
        column_data.append(cell_value)

    # Adiciona uma nova entrada ao dicionário, onde a chave é o valor do cabeçalho da coluna
    # e o valor é a lista com os valores da coluna
    data[header] = column_data

# Imprime o dicionário criado
print(data)

#### Credenciais da AWS
# aws_access_key_id = 'CHAVE_DE_ACESSO'
# aws_secret_access_key = 'CHAVE_SECRETA'

#### Cria um cliente para o serviço S3 da AWS
# s3 = boto3.client('s3', aws_access_key_id=aws_access_key_id, aws_secret_access_key=aws_secret_access_key)

#### Define o nome do bucket do S3 onde deseja armazenar as imagens
# bucket_name = 'NOME_DO_SEU_BUCKET'

# Itera pelas linhas do dicionário de dados
for row_index in range(num_rows - 1):
    
    # Verifica se há um link para a imagem na linha atual
    image_url_key = f'Foto Contenção 1_{row_index + 1}'
    print(image_url_key)
    if image_url_key in data:
        # Obtém o link para a imagem
        image_url = data[image_url_key][row_index]

        # Verifica se há uma descrição e um tipo para a imagem na linha atual
        image_description_key = f'DESCRICAO_DA_IMAGEM_{row_index + 1}'
        image_type_key = f'TIPO_DA_IMAGEM_{row_index + 1}'
        if image_description_key in data and image_type_key in data:
            # Obtém a descrição e o tipo da imagem
            image_description = data[image_description_key][row_index]
            image_type = data[image_type_key][row_index]

            # Define o nome do arquivo da imagem (você pode alterar isso como desejar)
            image_filename = f'{row_index}.{image_type}'

            # Faz o download da imagem a partir do link fornecido
            response = requests.get(image_url)

            #### Salva a imagem no bucket do S3
#             s3.put_object(Body=response.content, Bucket=bucket_name, Key=image_filename)

            #### Adiciona a descrição e o tipo da imagem como metadados do objeto no S3
#             s3.put_object_tagging(Bucket=bucket_name, Key=image_filename, Tagging={'TagSet': [{'Key': 'description', 'Value': image_description}, {'Key': 'type', 'Value': image_type}]})
```
