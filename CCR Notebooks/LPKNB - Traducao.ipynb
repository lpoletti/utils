# Funções para Tradução

```python
def traduzirTexto(item="Verifique a Informação Inserida"):
    for b,c in enumerate(traducao.items()):
        if item.lower() == c[0].lower():
            return c[1]
    return item
###tradução [{}] itera mais de uma vez pois uma vez só fica traduções sem fazer não encontrei o motivo ainda
def traducaoDefinitiva(item):
    for i in range(3):
        for k,a in enumerate(item):
            for ki,b in enumerate(a.items()):
                if b[0] in traducao:
                    t=traduzirTexto(b[0])
                    v=traduzirTexto(str(b[1]))
                    item[k][t] = a.pop(b[0])
                    item[k][t] = v
    return item
###tradução dict
def traduzirDict(item):
    value={}
    for k,v in item.items():
        key = traduzirTexto(str(k))
        va = traduzirTexto(str(v))
        value[key] = va
    return value
def traduzirTextoInverso(item="Verifique a Informação Inserida"):
    for b,c in enumerate(traducao.items()):
        if item.lower() == c[1].lower():
            return c[0]
    return item
```
