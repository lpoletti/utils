```python
import json
from helpers.fields import FeatureCollectionField
```

```python
uhit = Company.objects.get(name='GES - UHE-ESTREITO')
engie=User.objects.get(username='engie')
uhit
```

```python
folder = "Mosaico/"
filename = "mosaico"
filename = folder+filename
material = []
nome = 'Mosaico'
```

```python
field = FeatureCollectionField(
    required=False,
    allow_null=True,
    geometry_field="geometry",
    properties_field="properties",
)

with open(filename + '.geojson', 'r+') as f:
    material = json.load(f)
# print(material)
internal_values = field.to_internal_value(material)

shapefile = ShapeFile(
  created_by = engie,
  name = nome,
  description = nome,
#   parent = ShapeFile.objects.get(companies=uhit, name='PACUERA'),
  geometry = internal_values['geometry'],
  properties = internal_values['properties'],
  metadata = {
    "fill_color": "#faf8e1", # escolher cores
    "stroke_color": "#faf8e1" # escolher cores
}
)

shapefile.save()

shapefile.companies.add(uhit)
```
